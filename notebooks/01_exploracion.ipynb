```python
df_2023 = pd.read_excel('../data/raw/CONOSCE_ADJUDICACIONES2023_0.xlsx')
df_2024 = pd.read_excel('../data/raw/CONOSCE_ADJUDICACIONES2024_0.xlsx')
df_2025 = pd.read_excel('../data/raw/CONOSCE_ADJUDICACIONES2025_0.xlsx')

df_total = pd.concat([df, df_2023, df_2024, df_2025], ignore_index=True)

df_total.shape
```

```python

df_clean = df_total.dropna(subset=[
    'cantidad_adjudicado_item',
    'monto_referencial_item_soles',
    'monto_adjudicado_item_soles',
    'tipo_proveedor',
    'fecha_buenapro'
])

df_clean.shape
```

```python

df_clean['diferencia_monto'] = df_clean['monto_adjudicado_item_soles'] - df_clean['monto_referencial_item_soles']

df_clean['diferencia_pct'] = (df_clean['diferencia_monto'] / df_clean['monto_referencial_item_soles']) * 100

df_clean[['monto_referencial_item_soles', 'monto_adjudicado_item_soles', 'diferencia_monto', 'diferencia_pct']].describe()
```

```python
(df_clean['monto_referencial_item_soles'] == 0).sum()
```

```python
# Casos con monto referencial = 0 (posibles anomalías de registro)
casos_sin_referencial = df_clean[df_clean['monto_referencial_item_soles'] == 0]

# Dataset principal para clustering (sin estos casos extremos)
df_model = df_clean[df_clean['monto_referencial_item_soles'] != 0].copy()

print(f"Casos especiales (referencial = 0): {casos_sin_referencial.shape[0]}")
print(f"Dataset para modelo: {df_model.shape[0]}")
```

```python
df_model[['monto_referencial_item_soles', 'monto_adjudicado_item_soles', 'diferencia_monto', 'diferencia_pct']].describe()
```

```python
# Capear diferencia_pct entre -100% y +500% 
# (más allá de esto, consideramos "outlier extremo de registro")
df_model['diferencia_pct_cap'] = df_model['diferencia_pct'].clip(lower=-100, upper=500)

# Ver cuántos valores fueron afectados por el cap
afectados = ((df_model['diferencia_pct'] < -100) | (df_model['diferencia_pct'] > 500)).sum()
print(f"Registros afectados por el cap: {afectados}")

df_model['diferencia_pct_cap'].describe()
```

```python
# Convertir las columnas de fecha a tipo datetime (formato día/mes/año)
df_model['fecha_convocatoria'] = pd.to_datetime(df_model['fecha_convocatoria'], dayfirst=True)
df_model['fecha_buenapro'] = pd.to_datetime(df_model['fecha_buenapro'], dayfirst=True)

# Calcular días entre convocatoria y adjudicación
df_model['dias_proceso'] = (df_model['fecha_buenapro'] - df_model['fecha_convocatoria']).dt.days

# Ver estadísticas
df_model['dias_proceso'].describe()
```

```python
(df_model['dias_proceso'] < 0).sum()
```

```python
columnas_modelo = [
    'monto_referencial_item_soles',
    'monto_adjudicado_item_soles',
    'diferencia_monto',
    'diferencia_pct_cap',
    'dias_proceso'
]

df_features = df_model[columnas_modelo]

df_features.describe()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Escalar las variables
X_scaled = scaler.fit_transform(df_features)

# Convertir a DataFrame para verlo mejor
import numpy as np
X_scaled_df = pd.DataFrame(X_scaled, columns=columnas_modelo)

X_scaled_df.describe()
```

```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Tomar una muestra aleatoria (reproducible con random_state)
np.random.seed(42)
muestra_idx = np.random.choice(X_scaled.shape[0], size=2000, replace=False)
X_muestra = X_scaled[muestra_idx]

# Calcular el linkage (usando Ward, el más común)
Z = linkage(X_muestra, method='ward')

# Graficar el dendrograma
plt.figure(figsize=(12, 6))
dendrogram(Z, truncate_mode='lastp', p=30)
plt.title('Dendrograma - Contrataciones Públicas (muestra)')
plt.xlabel('Clusters')
plt.ylabel('Distancia')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Aplicar K-means con k=4
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
df_model['cluster'] = kmeans.fit_predict(X_scaled)

# Ver cuántos registros cayeron en cada cluster
df_model['cluster'].value_counts()
```

```python
# Promedio de cada variable, por cluster
df_model.groupby('cluster')[columnas_modelo].mean()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_model, 
    x='monto_referencial_item_soles', 
    y='diferencia_pct_cap', 
    hue='cluster', 
    palette='Set1',
    alpha=0.6
)
plt.xscale('log')  # escala logarítmica porque los montos varían muchísimo
plt.title('Clusters de Contrataciones según Monto y Diferencia %')
plt.xlabel('Monto Referencial (soles) - escala log')
plt.ylabel('Diferencia % (capeada)')
plt.show()
```

```python
# Nuevas variables de "comportamiento" (sin tamaño)
columnas_comportamiento = ['diferencia_pct_cap', 'dias_proceso']

X_comportamiento = df_model[columnas_comportamiento]

# Escalar
scaler2 = StandardScaler()
X_comp_scaled = scaler2.fit_transform(X_comportamiento)

# Método del codo: probar distintos k
from sklearn.cluster import KMeans

inercias = []
rango_k = range(1, 11)

for k in rango_k:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_comp_scaled)
    inercias.append(km.inertia_)

# Graficar el codo
plt.figure(figsize=(8, 5))
plt.plot(rango_k, inercias, marker='o')
plt.title('Método del Codo')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Inercia')
plt.xticks(rango_k)
plt.show()
```

```python
# K-means con las variables de comportamiento, k=4
kmeans_comp = KMeans(n_clusters=4, random_state=42, n_init=10)
df_model['cluster_comportamiento'] = kmeans_comp.fit_predict(X_comp_scaled)

# Tamaño de cada cluster
print(df_model['cluster_comportamiento'].value_counts())

# Características de cada cluster
df_model.groupby('cluster_comportamiento')[columnas_comportamiento].mean()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_model,
    x='dias_proceso',
    y='diferencia_pct_cap',
    hue='cluster_comportamiento',
    palette='Set1',
    alpha=0.5
)
plt.title('Clusters de Comportamiento: Días de Proceso vs Diferencia %')
plt.xlabel('Días de Proceso')
plt.ylabel('Diferencia % (capeada)')
plt.show()
```
